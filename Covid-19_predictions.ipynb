```python
import pandas as pd
import matplotlib as mplib
import matplotlib.pyplot as plt

```

```python
#get the dataset

covid_confirmed_df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv')
covid_death_df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Deaths.csv')
covid_recovered_df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Recovered.csv')
```

```python
covid_confirmed_df.head()
```

```python
covid_death_df.head()
```

```python
covid_recovered_df.head(10)
```

```python
covid_confirmed_df = covid_confirmed_df.melt(id_vars=['Province/State','Country/Region','Lat','Long'])
```

```python
covid_confirmed_df.head(10)
```

```python
covid_confirmed_df = covid_confirmed_df.rename({'variable':'Date','value':'Confirmed Cases'}, axis = 'columns')
```

```python
covid_confirmed_df.head()
```

```python
covid_death_df = covid_death_df.melt(id_vars=['Province/State','Country/Region','Lat','Long'])
covid_death_df = covid_death_df.rename({'variable':'Date','value':'Deaths'}, axis = 'columns')
covid_death_df.head()
```

```python
covid_recovered_df = covid_recovered_df.melt(id_vars=['Province/State','Country/Region','Lat','Long']).rename({'variable':'Date','value':'Recovered Cases'}, axis = 'columns')
covid_death_df.head()
```

```python
combined_df = [covid_confirmed_df,covid_death_df,covid_recovered_df]
combined_df
```

```python
combined_df = [covid_confirmed_df, covid_death_df, covid_recovered_df]
combined_df = [df.set_index(['Province/State','Country/Region','Lat','Long','Date']) for df in combined_df]
combined_df = combined_df[0].join(combined_df[1:])
combined_df.head()
```

```python
combined_df = combined_df.reset_index()
combined_df.head()
```

```python
combined_df[['Lat','Long','Confirmed Cases','Deaths','Recovered Cases']] = combined_df[['Lat','Long','Confirmed Cases','Deaths','Recovered Cases']].apply(pd.to_numeric)
```

```python
combined_df.head()
```

```python
combined_df[['Date']] = combined_df[['Date']].apply(pd.to_datetime)
combined_df.head()
```

```python
combined_df.dtypes
```

```python
combined_df = combined_df.rename({'Province/State':'State','Country/Region':'Country'}, axis = 'columns')
combined_df.head()
```

```python
combined_df.query("State == 'New York' & Country=='US'")
```

```python
combined_df['Active'] = combined_df['Confirmed Cases'] - combined_df['Deaths'] - combined_df['Recovered Cases']
combined_df.head()
```

```python
max_df = combined_df[combined_df.Date == combined_df['Date'].max()]
```

```python
max_df[['Country','Active']].sort_values('Active',ascending=False).head(10)
```

```python
max_df[['Country','Confirmed Cases']].groupby(['Country']).sum().sort_values('Confirmed Cases',ascending = False).head(10)
```

```python
mplib.rcParams['figure.figsize'] = (20, 10)
mplib.rcParams['axes.grid'] = False
combined_df[['Date','Country','State','Active']].query("Country=='China' & State=='Hubei'").plot(x='Date',
 y='Active',kind='bar', title="Corona virus Trend for Country China and State Hubei")
```

```python
mplib.rcParams['figure.figsize'] = (20,10)
mplib.rcParams['axes.grid'] = False
combined_df[['Date','Country','State','Confirmed Cases']].query("Country == 'Iran'").plot(x='Date',
   y='Confirmed Cases',kind='line',title='Corona Virus Spread in India')
```

```python
#Corona Virus Cases plotting in India

mplib.rcParams['figure.figsize'] = (20,10)
mplib.rcParams['axes.grid'] = False
combined_df[['Date','Country','State','Confirmed Cases','Deaths','Recovered Cases']].query("Country == 'India'").plot(x='Date',
        y=['Confirmed Cases','Deaths','Recovered Cases'],kind='line',title='Corona Virus Spread in India')
```

```python
mplib.rcParams['figure.figsize'] = (20,10)
mplib.rcParams['axes.grid'] = False
combined_df[['Date','Country','State','Confirmed Cases','Deaths','Recovered Cases']].query("Country == 'United Kingdom'").plot(x='Date',
        y=['Confirmed Cases','Deaths','Recovered Cases'],kind='line',title='Corona Virus Spread in UK')
```
